```python
number = int(input("ENTER THE NUMBER"))
sum = 0
while(number != 0):
    sum = sum + (number % 10)
    number = number // 10

print("sum of the digits: ",sum)
```

```python
#Armstrong number
number = int(input("Enter a number: "))
num = number
power = len(str(number))
sum = 0
while(number != 0):
    sum += ((number % 10) ** power)
    number = number // 10

if(num == sum):
    print("Armstrong")
else:
    print("not Armstrong")
```
