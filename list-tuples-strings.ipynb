```python
a = [10, 20, 30]
b = [10, 20, 30]

c = a
d = b.copy() #this creates a new memory address for d, 
'''this creates a new copy for b(200) value in the bracket is the memory address, therefore d(300) will be created when the copy is done'''

print(a == b)
print(a is b)
print(b == d)
print(b is d) # this means we compare different addresses, thus we get false
print(a == c)
print(a is c) #here we get answer as true, as we copy the same object of a into c


#printing the memory addresses
print(id(a))
print(id(b))
print(id(c))
print(id(d))
```
