```python
#ask the user selling price and the discount
'''format of the output should be 
selling price:  1000
discount: 20%
total: 800
GST: 18%
print: total'''


selling_price = int(input("enter the selling price: "))
discount = int(input("Enter the discount: "))

total_after_dist = selling_price - ((selling_price * (discount / 100)))
total_price = (total_after_dist + (0.18 * total_after_dist))


print("Selling price : ",selling_price)
print("Discount : ",discount)
print("----------------------------")
print("total : ",total_after_dist)
print("GST : 18%")
print("----------------------------")
print("final total: ",total_price)

```

```python
# ASSIGNMENT OPERATOR
# = += -= *= /= //= **=
a = 10
b = 20

a += b   # a = a + b
print(a)

a -= b   # a = a - b
print(a)

a /= b
print(a)  # a = a / b 
```

```python
#BITWISE OPERATOR
# << Shift left
# >> Shift right 
# &  bitwise and
# |  bitwise or
# ^  bitwise xor
# ~ bitwise complement
```

```python
print(50 & 40)
print(20 >> 2)
print(20 << 2)
print(50 ^ 40)
print(50 | 40)
print(~9)
```
