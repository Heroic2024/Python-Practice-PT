```python
# Passing the list as function and finding the 2nd largest element without sorting
def fun(ff):
    largest = ff[0]
    secondlargest = ff[0]

    for i in ff:
        if i > largest:
            secondlargest = largest
            largest = i
        elif i > secondlargest and i!= largest:
            secondlargest = i

    print(secondlargest)

l = [78, 34, 23, 45, 78]
fun(l)
print()

```

```python
#Rearrage positive and negative numbers:
#Question: given an array containing both posotive and negative number, reaarange them in an alternating fashion
#Logic: Seprate positive and negative numbers, then merge them alternatively
#Sample input: [-1,2,-3,4,5,-6] Output: [-1,2,-3,4,-6,5]
# Rearrange positive and negative numbers alternatively
def fun1(f):
    positive = []
    negative = []

    for i in f:
        if i >= 0:
            positive.append(i)
        else:
            negative.append(i)
    result = []
    i = 0
    while i < len(positive) and i < len(negative):
        result.append(negative[i])
        result.append(positive[i])
        i += 1
    return result
f = [-1, 2, -3, 4, 5, -6]
fun(f)
print()

```

```python
#for palindromic list
#write a fuction to check if a list is a palindrome.
# Sample input: [1,2,3,2,1] Output: Palindrome
# Function to check if a list is palindrome

def fun3(l):
    n = len(l)
    for i in range(n // 2):
        if l[i] != l[(n-i)-1]:
            print("Not Palindrome")
            return
    print("Palindrome")
l = [1,2,3,2,1]
fun3(l)
print()

```

```python
#Product of array except self:
#Question: given an array, return an array where each element is the product of all the elements in the array except itself
#Sample input: [1,2,3,4] Output: [24,12,8,6]
def fun4(l):
    product = 1
    for i in l:
        product *= i
        result = []
    for i in l:
        result.append(product // i)
    return result
    
l = [1,2,3,4]
fun4(l)


```

```python
#Move zeros to the end:
#Question: Given an array, move all the zeros to the end chaning the order of non-zero elements.
# Move zeros to the end
def fun5(arr):
    result = []
    count = 0

    for i in arr:
        if i != 0:
            result.append(i)
        else:
            count += 1
    for i in range(count):
        result.append(0)
    print(result)
a = [0,1,0,3,12]
fun5(a)
print()


```

```python
#Dutch National flag problem:
#Question: Given an array containing only 0s, 1s and 2s, sort the array in a single pass, without using sort function
#Sample input: [2,0,1,2,1,0] Output:[0,0,1,1,2,2]
l = [2,0,1,2,1,0]
def fun6(l):
    low = 0
    mid = 0
    high = len(l) - 1
    while mid <= high:
        if l[mid] == 0:
            l[low], l[mid] = l[mid], l[low]
            low += 1
            mid += 1
        elif l[mid] == 1:
            mid += 1
        else:
            l[mid], l[high] = l[high], l[mid]
            high -= 1 
    return l
print(fun6(l))
print()

```
