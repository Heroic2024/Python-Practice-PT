```python
a = {1,2,3}
b = {4,5,6,2}
print(a.union(b)) # prints unique values
print(a|b) # same as union()
print(a.intersection(b)) # prints common values
print(a&b) # same as intersection()

print(a - b) # does not print the common element and prints the other elements of the first set (prints a in a - b)
```

```python
a = {1,2,3,4,5,6}
b = {3,4,5}
print(a.issubset(b))
print(b.issubset(a))

print(a.issuperset(b))
print(a.isdisjoint(b)) # if no elements are common, then print true
print(a.symmetric_difference(b))
```

```python
#BUBBLE SORT 
#find if the current element is greater than the next element, and then swap the elements

l = [10,3,20,30,23,45]
for i in range(0, len(l)-1):
    if(l[i] > l[i+1]):
        l[i],l[i+1] = l[i+1],l[i]

print(l)
```

### DICTIONARY

```python
#dictionary
#{key:value}

#values are mutable
#keys are immutable

d = dict()
print(type(d))

student = {
    "name": "John",
    "rollno": 28,
    "age":20,
    "subjects": "ml",
}
print(student)
print(student["name"])


#id
d = {"name": "John", id: 20}
print(d)
print(d[id])

```

```python
student = {"name":"Raj"}
print(student)
student["rollno"] = 28
print(student)
```

```python
stud = {
    "roll":12,
    "name":"Aryan",
    "age":20,
    "subjects": ["python", "java", "c++"],
    "city":"Los Angeles"
}

if "name" in stud:
    print("yes")
else:
    print("no")

#if tried searching for a key that is not present, it will give an error
if 12 in stud:
    print("yes")
else:
    print("no")

# function to check the keys
if "name" in stud.keys():
    print("yes")
else:
    print("no")
```

```python
stud = {
    "roll":12,
    "name":"Aryan",
    "age":20,
    "subjects": ["python", "java", "c++"],
    "city":"Los Angeles"
}

for i in stud:
    print(i, end = " ") # prints the keys
print()
for i in stud.keys():
    print(i, end = " ") # prints the keys
print()
for i in stud.values():
    print(i, end = " ") # prints the values
print()
for i in stud.items():
    print(i, end = " ") # prints the key and value as a tuple
```

```python
stud = {
    "roll":12,
    "name":"Aryan",
    "age":20,
    "subjects": ["python", "java", "c++"],
    "city":"Los Angeles"
}
x = stud.pop("city") # removes the key value pair with the key "city" and returns the value of the removed key
print(stud)
print(x)

x = stud.popitem() # removes the last key value pair and returns it as a tuple
print(x)
print(stud)
```

```python
stud = {
    "roll":12,
    "name":"Aryan",
    "age":20,
    "subjects": ["python", "java", "c++"],
    "city":"Los Angeles"
}
x = {"city":"Los Angeles","gender":"male"}
stud.update(x) # updates the dictionary with the key value pairs from x
print(stud)
y = {"roll":45}
stud.update(y) # updates the value of the key "roll" to 45
print(stud)
```

```python
student = dict.fromkeys(["roll","name","city"],55) # creates a dictionary with the keys from the list and the value as 55 for all keys
print(student)


student = dict.fromkeys("abc",55) # creates a dictionary with the keys from the string and the value as 55 for all keys
print(student)

student = dict.fromkeys(range(0,5),55) # creates a dictionary with the keys from the range and the value as 55 for all keys
print(student)
```

```python
student =  {"a":1}
val = student.setdefault("b", 2) # if the key is present, then it returns the value of the key, otherwise it adds the key with the value and then returns the value
print(student)
print(val)
```

```python
stud = {
    "roll":12,
    "name":"Aryan",
    "age":20,
    "subjects": ["python", "java", "c++"],
}

print(stud.get("name")) # returns the value of the key "name"
print(stud.get("city")) # returns None as the key is not present
print(stud.get("city", "Not Found")) # returns "Not Found" as the key is not present
```

```python
#print only the ones whose names start with a, print name as key and value as the length of the name
l = ["amit","anjali","rohit","yash"]
dict = {i:len(i) for i in l if i[0] == 'a' or i[0] == 'A'}
print(dict)
```

```python
#print value as odd or even
l = [1,2,3,4,5]
dict = {i:"odd" if i % 2 != 0 else "even" for i in l}
print(dict)
```

```python
#print values as zero, positive or negative
n = [-1,-2,3,0,9,-11,12]
dict = {i:"zero" if i == 0 else "positive" if i > 0 else "negative" for i in n}
print(dict)
```
