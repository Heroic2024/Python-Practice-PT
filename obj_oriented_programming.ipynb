class is a keyword which is used to create user defined datatype



```python
class Student:
    def accept(self, name, roll):
        self.name = name
        self.roll = roll

    def display(self):  #methods also known as member functions
        print("Name: ", self.name)
        print("Roll: ", self.roll)

s1 = Student()
s1.accept("John", 101)
s1.display()
print(s1.name)
print(s1.roll)
```

```python
class Student:
    clg = "NMIMS"
    def __init__(self, roll, name):
        self.roll = roll
        self.name = name
    
    
s1 = Student(101, "Bob")
s2 = Student(102, "Charlie")
s3 = Student(103, "David")

print(Student.clg) #should be printed with class name usually, can also be printed with object name but not recommended
print(s1.roll, s1.name, s1.clg)
print(s2.roll, s2.name, s2.clg)
print(s3.roll, s3.name, s3.clg)
    
s1.clg = "IIT" # this will create a new variable clg for s1 object and will not change the class variable clg
print(Student.clg) # will print NMIMS
print(s1.clg) # will print IIT


Student.clg = "MIT"
print(Student.clg) # will print MIT
print(s1.clg) # will print IIT because s1 has its own clg variable
```

```python
class Student:
    def __init__(self, roll, name):
        self.roll = roll
        self.name = name
    
    @staticmethod
    def fun():
        print("This is a static method")

    def fun1(self):
        print("This is a non-static method ")

s1 = Student(101, "Eve")
s1.fun() # can be called with object name but not recommended
Student.fun() # should be called with class name
s1.fun1() # can be called with object name
```

Polymorphism

1. Compile Time
- method overloading
- operator overloading

2. Run time Polymorphism
- method overloading

- Note: in a class can have exactly same name and diff parameter method called method overloading

```python
class Test:
    def fun(self, a):
        print("from fun 1 parameter")

    def fun(self,a,b):
        print("from fun 2 parameter")

    
t = Test() 
print(t.fun(1)) #this gives an error 
print(t.fun(1,2))

```

```python
# to avoid the above case, we use the same function for different number of parameters
# we declare one variable as 0 and check if there is no value passed for that parameter
class Test:
    def add(self,a,b,c = 0):
        if c == 0:
            print("addition of 2 nums",a+b)
        else:
            print("addition of 3 numbers",a+b+c)


t = Test()
t.add(1,2,3)
t.add(1,2)
```

- Constructor

```python
# if a constructor is not initialized, it is by default done when a class is created
class Test:
    def __init__(self):
        print("hello from default")
    
    def __init__(self,a,b):
        print("from parameterized constructor..")

t = Test(1,2) # this will give an error because the default constructor is overridden by the parameterized constructor and there is no default constructor available now, python does not support constructor overloading like java, we can only have one constructor in a class, if we declare another constructor, the previous one will be overridden and will not be available now, to avoid this we can use the same constructor for different number of parameters by declaring one variable as 0 and checking if there is no value passed for that parameter

```

```python
class Test1:
    def add(self,a,b):
        print("addition of 2 nums",a+b)
    
class Test2(Test1):
    def add(self,a,b,c):
        print("addition of 3 nums",a+b+c)

t = Test2()
#t.add(1,2) # this will give an error because the add method in Test2 class is overriding the add method in Test1 class and it is expecting 3 parameters but we are passing only 2 parameters, to avoid this we can use the same function for different number of parameters by declaring one variable as 0 and checking if there is no value passed for that parameter
t.add(1,2,3)
```

- method overriding 

both the class having parent child relationship and have exactly same name and same parameter method called

```python
class Parent:
    def car(self):
        print("bmw car...")

class Child(Parent):
    pass

c1 = Child()
c1.car()
```

```python
class Parent:
    def car(self):
        print("bmw car...")

class Child(Parent):
    def car(self):
        print("mercedes car...")

c1 = Child()
c1.car()
```
