class is a keyword which is used to create user defined datatype



```python
class Student:
    def accept(self, name, roll):
        self.name = name
        self.roll = roll

    def display(self):  #methods also known as member functions
        print("Name: ", self.name)
        print("Roll: ", self.roll)

s1 = Student()
s1.accept("John", 101)
s1.display()
print(s1.name)
print(s1.roll)
```

two types: Default and parameterized

```python
class Student:
    def __init__(self, name, age):
        self.name = name
        self.age = age

    def display(self):
        print("Name:",self.name)
        print("Age:",self.age)
    
s1 = Student("Alice", 20)
s1.display()
```

Static variable and static method 

static varialb and method: only single copy exists through out the program

called by class 

```python
class Student:
    clg = "NMIMS"
    def __init__(self, roll, name):
        self.roll = roll
        self.name = name
    
    
s1 = Student(101, "Bob")
s2 = Student(102, "Charlie")
s3 = Student(103, "David")

print(Student.clg) #should be printed with class name usually, can also be printed with object name but not recommended
print(s1.roll, s1.name, s1.clg)
print(s2.roll, s2.name, s2.clg)
print(s3.roll, s3.name, s3.clg)
    
s1.clg = "IIT" # this will create a new variable clg for s1 object and will not change the class variable clg
print(Student.clg) # will print NMIMS
print(s1.clg) # will print IIT


Student.clg = "MIT"
print(Student.clg) # will print MIT
print(s1.clg) # will print IIT because s1 has its own clg variable
```

```python
class Student:
    def __init__(self, roll, name):
        self.roll = roll
        self.name = name
    
    @staticmethod
    def fun():
        print("This is a static method")

    def fun1(self):
        print("This is a non-static method ")

s1 = Student(101, "Eve")
s1.fun() # can be called with object name but not recommended
Student.fun() # should be called with class name
s1.fun1() # can be called with object name
```

Inheritance 

```python
class Parent:
    def fun1(self):
        print("This is a method of Parent class")

class Child:
    pass

c1 = Child()
c1.fun1() # will give error because Child class does not have fun1 method and it is not inheriting from Parent class
```

```python
class Parent:
    def fun1(self):
        print("This is a method of Parent class")

class Child:
    pass

c1 = Child()
c1.fun1() # will give error because Child class does not have fun1 method and it is not inheriting from Parent class
```

```python
class Parent:
    def fun1(self):
        print("This is a method of Parent class")

class Child(Parent):
    pass

c1 = Child()
c1.fun1() # will give error because Child class does not have fun1 method and it is not inheriting from Parent class
```
