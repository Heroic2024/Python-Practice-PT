```python
# types of functions

# 1. functions with no arguments and no return type
def functionname():
    print("Aryan Mhatre")

functionname()

# 2. functions with arguments and no return type
def add(a,b,c):
    sum = a + b + c
    print(sum)
add(10,20,30)


# 3. function with no arguments and return type
def add():
    a = 12
    b = 13
    return a + b

print(add())

# 4. function with arguments and return type
def add(a,b):
    return a + b

print(add(10,20))

# 5. function with default arguments
def add(a,b,c = 0):
    print(a + b + c)

add(10,20) # c will take default value 0
add(10,20,30) # c will take value 30


# 6. functions with variable argument
def fun(*a): # a is a tuple and return type is tuple
    print(a)

fun(10)
fun(10,20)

def fun(*a,b,c):
    print(a) # a is a tuple
    print(b) # b is an integer
    print(c) # c is an integer

# fun(10,20,30,40,50) #this will give error because we have not provided value for b and c

def fun1(a,b,*c):
    print(a) # a is an integer
    print(b) # b is an integer
    print(c) # c is a tuple

fun1(10,20,30,40,50)  # 10 will be assigned to a, 20 will be assigned to b and (30,40,50) will be assigned to c
```

```python
#7. function with variable keyword arguments
# input(key value)

def fun(**a): # a is a dictionary and return type is dictionary
    print(a)

fun(x = 1,y = 2,z = 3)
fun(x = 1,y = 2,z = 3,w = 90)


#8. function with keyword only arguments
def fun(*,a,b,c):
    print(a) # a is an integer
    print(b) # b is an integer
    print(c) # c is an integer

# NAME NEEDS TO BE GIVEN TO VARIABLES AFTER *


# fun(1,2,3) # this will give error because we have not provided value for a,b and c as keyword arguments

fun(a = 1,b = 2,c = 3) # this will work because we have provided value for a,b and c as keyword arguments
fun(c = 3,b = 2,a = 1) # this will also work because we have provided value for a,b and c as keyword arguments
```

```python
def fun(x,*,y):
    print(x) # x is an integer
    print(y) # y is an integer

# fun(10,20) # this will give error because we have not provided value for y as keyword argument
fun(10,y = 20) # this will work because we have provided value for y as keyword argument
fun(x = 10,y = 20) # this will also work because we have provided value for y as keyword argument
```

```python
# 9. positional only arguments
def fun(a,/,b):
    print(a) # a is an integer
    print(b) # b is an integer

fun(10,20) # you don't give name of a
#fun(a = 10,b = 20) # this will give error because a is a positional only argument
fun(10,b = 20) 


def fun1(a,b,/):
    print(a) # a is an integer
    print(b) # b is an integer

fun1(10,20) # you don't give name of a
```

```python
def fun(a,/,b,*,c):
    print(a) # a is an integer
    print(b) # b is an integer
    print(c) # c is an integer

fun(10,20,c = 30) # this will work because we have provided value for c as keyword argument
fun(10,b = 20,c = 30) # this will also work because we have provided value for c as keyword argument
```

```python
# function replacement 

def add(a,b):
    print(a + b)

def add(a,b,c):
    print(a + b + c)

add(1,2,3)
add(1,2) # this will give error because we have defined a new function with same name and it will replace the old function

```

```python
# local and global
def fun1(x):
    print(x) # x is a local variable

def fun2():
    print(x) # this will give error because x is a local variable and it is not defined in this function

x = 20
fun1(x) # this will work because x is a global variable and it is defined outside the function
fun2() # this will work because x is a global variable and it is defined outside the function
```
