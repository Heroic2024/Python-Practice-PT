#### Strings

```python
# Strings are immutable
#indexing is possible, and slicing can be done
s = "hello"
print(s)
s = "abc"
print(s)
```

```python
s = "hello"
print(s[0])
print(s[-1])
print(s[-2])
```

#### IMPORTANT: here we can see that the both s1 and s2 have the same address indicating, s1 and s2 both point to the same memory of the string "hello"


```python
# String Concatenation
s1 = "hello"
s2 = "hello"
s3 = s1 + s2
print(s3)

print(id(s1))
print(id(s2))
print(id(s3))
```

### Here we can see the object being modified and thus a new memory address is generated

```python
s1 = "hello"
print(s1)
print(id(s1))
s1 = s1 + "abc"
print(s1)
print(id(s1))
```

```python
#count vowels in a string

s = "aadit"
vow_cnt = 0
for i in range (0,len(s)):
    if(s[i] in "aeiouAEIOU"):
        vow_cnt += 1

print("number of vowels in the string are: ",vow_cnt)
```

```python
#check string is a palindrome
s = "racecar"
s_rev = s[::-1]
if(s == s_rev):
    print("Palindrome")
else:
    print("not Palindrome")
```

```python
# remove duplicate characters
s = "racecar"
s1 = ""
for i in s:
    if i not in s1:
        s1 += i

print(s1)
```
