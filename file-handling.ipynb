```python
# Write mode
file = open("sample-text.txt","w")
file.write("Hello world \n")
file.close()
```

```python
# Append mode
file = open("sample-text.txt","a")
file.write("Appending a new line \n")
file.close()
```

```python
# Read mode
file = open("sample-text.txt","r") 
print(file.read())
file.close()
```
