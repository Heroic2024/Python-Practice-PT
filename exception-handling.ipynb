```python
# custom exception class
class PercentGreaterThan100(Exception):
    '''this is an exception class'''
    pass

per = float(input("Enter per:  "))
try:
    if per > 100:
        raise PercentGreaterThan100("percentage cannot be greater than 100")
except PercentGreaterThan100 as e:
        print(e)
        print(e.__class__,e)
print("hello")
```

### lambda functions

```python
# lambda functions

def sqr(x):
    return x*x

sqr(5)
```

```python
sqr = lambda x: x*x
print(sqr(5))
```

```python
cube = lambda x: x*x*x
print(cube(5))
```

```python
even = lambda x:x%2==0
print(even(2))
```

```python
# comparing 2 numbers
x = lambda x,y: x if x>y else y
print(x(10,20))

x = lambda x,y: print( x if x>y else y)
x(10,30)
```

### map, filter, reduce, enumerate

```python
# map, filter, reduce, enumerate
# passing function as an argument

# map()
l = [1,2,3,4,5]
def fun(l):
    for i in l:
        l.append(i+1)

a =list(map(lambda x:x+1,l)) #map function takes a function and an iterable as an argument and returns a map object which is an iterator that applies the function to every item of the iterable
print(a)
```

```python
l = [1,2,3,4,5]
s = list(map(lambda x:x**2,l))
print(s)
print(list(map(lambda x:x**3,l)))
```

```python
l = ["10","20","30"]
s = list(map(int,l))
print(s)
```

```python
l = ["hello","ram","raj"]
s = list(map(lambda x:x.capitalize(),l))
print(s)
```
